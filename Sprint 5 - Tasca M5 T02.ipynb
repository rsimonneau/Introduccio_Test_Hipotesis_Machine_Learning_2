### Exercici 1

Parteix el conjunt de dades adjunt en train i test. Estudia els dos conjunts per separat, a nivell descriptiu.

També adjunt trobaràs una descripció de les diferents variables del dataset.

```python
import pandas as pd

pd.read_csv('housing_data.csv')
```

```python
# Generem el data frame amb el nom de les columnes

columnes = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']

Boston_df = pd.read_csv('housing_data.csv', names = columnes)

Boston_df
```

```python
import numpy
import matplotlib.pyplot as plt

# Prenem les variables següents:
    # 'RM': average number of rooms per dwelling (mitjana habitacions per habitatge)
    # 'MEDV': Median value of owner-occupied homes in $1000's (valor mitjà dels habitatges en milers de dòlars)

x = Boston_df['RM']
y = Boston_df['MEDV']

# Gràfic

plt.scatter(x, y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x = Boston_df['RM']
y = Boston_df['MEDV']

# utilitzem la funcio 'train_test_split' per fer la partició del conjunt de dades. 
    # x: dades (atributs)
    # y: valors que es volen predir (target)
    # test_size: proporció de dades que s'assigna al conjunt 'TEST' (en aquest cas 20%) - El 80% restant s'assigna al conjunt 'TRAIN'
    # random_state: es recomana utilitzar un valor fixe a l'hora de realitzar la partició en 'train' i 'test'. El valor fixe permet que la partició sempre sigui la mateixa.
      # (s'acostuma a utilitzar el 'random_state = 42')

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Estudiem el conjunt 'TRAIN' a nivell descriptiu 

print(x_train.describe())

print(y_train.describe())

# Observem que s'analitzen 404 valors (el 80% de les 506 files)
```

```python
# Estudiem el conjunt 'TEST' a nivell descriptiu

print(x_test.describe())

print(y_test.describe())

# Observem que s'analitzen 102 valors (el 20% de les 506 files)
```

```python
# Fem la gràfica de la partició 'TRAIN'

plt.scatter(x_train, y_train)
plt.show()

# Similar a la gràfica de totes les dades (abans de la partició)
```

```python
# Fem la gràfica de la partició 'TEST'

plt.scatter(x_test, y_test)
plt.show()
```

```python
# Tracem la línia de 'Regressió Polinomial' en la partició 'TRAIN'

Model = numpy.poly1d(numpy.polyfit(x_train, y_train, 2)) # en aquest cas s'ajusta la de 2on grau '2'

# ajust de la linia:

Linia = numpy.linspace(4, 9, 100)

# Gràfica:

plt.scatter(x_train, y_train)
plt.plot(Linia, Model(Linia))
plt.show()
```

```python
# Calculem la R2 de la partició 'TRAIN'

    # La R2 (R-Quadrat), medeix la relació entre l'eix 'X' i l'eix 'Y', i varia entre 0 i 1, on 0 siginifica que no hi ha relació i 1 significa que estan totalment relacionades.

# importem r2_score() de la llibreria 'sklearn'

from sklearn.metrics import r2_score

Model = numpy.poly1d(numpy.polyfit(x_train, y_train, 2))

r2 = r2_score(y_train, Model(x_train))

print(r2)
```

```python
# Calculem igualment la R2 de la partició 'TEST'

    # Podem comprovar si en el  model 'TEST' obtenim el mateix resultat que al model 'TRAIN'

from sklearn.metrics import r2_score

Model = numpy.poly1d(numpy.polyfit(x_train, y_train, 2))

r2 = r2_score(y_test, Model(x_test))

print(r2)
```

Análisi R2
- Havent calculat la R2 en el model 'TRAIN' veiem que el valor se situa mes a prop del 1 que del 0. Tot i que encara està allunyat del 1, podem dir que existeix certa relació entre 'x' i 'y'
- Que el valor de R2 en 'TEST' sigui inferior al model de R2 en 'TRAIN' pot significar que el model no estigui prou ajustat. En aquest cas la diferencia és petita, així doncs podria utilitzar-se el model per predir valors, tot i que amb análisi posterior dels resultats. 

```python
# Mirem de predir algun valor:

print(Model(7))

# Obtenim el valor mitjà de un habitatge amb 7 habitacions
```

```python
print(Model(9))

# Obtenim el valor mitjà de un habitatge amb 9 habitacions.

    # En aquest cas, caldria analitzar en profunditat si valor mitjà arriba a un màxim per a habitatges de 8 habitacions.
```

```python
print(Model(6))

# Obtenim el valor mitjà de un habitatge amb 6 habitacions.
```

```python
# desviació estandard de les dades estandaritzades

scaler.scale_
```

- Creació de columnes dummies

```python
# Creem columnes 'dummies' de la variable 'RAD' 

dummies = pd.get_dummies(Boston_df['RAD'], prefix='RAD')

# Unim les columnes dummies al DataFrame original (Boston_df)

Boston_df_dummy = pd.concat([Boston_df, dummies], axis = 1)

Boston_df_dummy
```

---------------------------------------------------------------------------------------------------------------

- Polinomis

```python
import numpy as np

# Creem característiques polinomials de la variable 'RM'

    # Primer generem la array de la variable 'RM'
    
RM_array = np.array(Boston_df['RM']).reshape(-1, 1)

RM_array
```

```python
# Generem les característiques polinomials

from sklearn.preprocessing import PolynomialFeatures

polinomi = PolynomialFeatures(degree=2) # En aquest cas de 2 grau (1, x, x^2)

# Generem la array amb la columna original i les característiques polinomials agregades:

Array_polinomial = polinomi.fit_transform(RM_array)

Array_polinomial
```

```python
# Agreguem les noves columnes generades al DataFrame inicial 'Boston_df'

    # Primer generem el DataFrame de les noves columnes generades

Boston_df_polinomial = pd.DataFrame(Array_polinomial, columns=['RM^0','RM^1','RM^2'])

Boston_df_polinomial

    # En segon lloc, agreguem les columnes al DataFrame original, ontenint un nou DataFrame 'Boston_df_new'

Boston_df_new = pd.concat([Boston_df, Boston_df_polinomial, dummies], axis = 1) # afegim també les columnes dummmies generades anteriorment

Boston_df_new
```
